```python
import pandas as pd
import yfinance as yf
import numpy as np
from scipy.optimize import minimize
# List of ETFs
etfs = ["QQQ", "VBR", "SPDW", "VWO", "REET", "BND", "EMB", "JPST", "GLD", "WTMF"]

# Set the start and end dates for a 5-year window ending in November 2023
end_date = "2023-11-30"
start_date = "2018-12-01"

# Create an empty DataFrame to store the adjusted close prices
all_data = pd.DataFrame()

for etf in etfs:
    # Download the data
    data = yf.download(etf, start=start_date, end=end_date, progress=False)

    # Resample the data to get monthly frequency using the adjusted close prices
    monthly_data = data["Adj Close"].resample("M").last()

    # Add the data to the main DataFrame
    all_data[etf] = monthly_data

# Save the DataFrame to a CSV file
all_data.to_csv("etf_prices.csv")

print("Data downloaded and saved to etf_prices.csv")

```

```python
# Calculate monthly returns
monthly_returns = etf_data.pct_change().dropna()
```

```python
# Annualize the statistics (mean and standard deviation)
annualized_mean = monthly_returns.mean() * 12
annualized_std = monthly_returns.std() * np.sqrt(12)
print(annualized_mean)
print(annualized_std)
```

```python
# Calculate the covariance matrix
covariance_matrix = monthly_returns.cov()
```

```python
# Calculate the correlation matrix
correlation_matrix = monthly_returns.corr()
```

```python
# Number of assets
n_assets = len(etfs)

# Equal weights for each ETF (1/N portfolio)
equal_weights = np.array([1/n_assets] * n_assets)
```

```python
# Define functions for portfolio statistics
def portfolio_mean(weights, mean_returns):
    return np.sum(mean_returns * weights)

def portfolio_std(weights, covariance_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(covariance_matrix, weights)))

def portfolio_variance(weights, covariance_matrix):
    return np.dot(weights.T, np.dot(covariance_matrix, weights))
```

```python
# Constraints for optimization (weights sum to 1 and are non-negative)
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1},
               {'type': 'ineq', 'fun': lambda x: x})
```

```python
# Boundaries for the weights (between 0 and 1)
bounds = tuple((0, 1) for asset in range(n_assets))
```

```python
# Initial guess (equal weights)
initial_guess = equal_weights
```

```python
# Minimize the portfolio variance
optimal_variance_result = minimize(portfolio_variance, initial_guess, args=(covariance_matrix,),
                                   method='SLSQP', bounds=bounds, constraints=constraints)
```

```python
#PART 3
annualized_mean, annualized_std, correlation_matrix, optimal_variance_result
import matplotlib.pyplot as plt
```

```python
# Function to minimize (for a given target return), now correctly accepting covariance matrix
def min_variance_for_target_return(weights, covariance_matrix):
    return portfolio_variance(weights, covariance_matrix)
```

```python
# Define the range of target returns
target_returns = np.linspace(annualized_mean.min(), annualized_mean.max(), 50)
```

```python
# Initialize lists to store results
portfolio_stds = []
portfolio_means = []

# Run optimization for each target return
for target_return in target_returns:
    constraints_with_return = constraints + (
        {'type': 'eq', 'fun': lambda x: portfolio_mean(x, annualized_mean) - target_return},)
    result = minimize(min_variance_for_target_return, initial_guess, args=(covariance_matrix,),
                      method='SLSQP', bounds=bounds, constraints=constraints_with_return)
    portfolio_stds.append(portfolio_std(result.x, covariance_matrix))
    portfolio_means.append(target_return)
```

```python
# Plotting the Efficient Frontier
plt.figure(figsize=(10, 6))
plt.scatter(portfolio_stds, portfolio_means, c=portfolio_means / np.array(portfolio_stds), cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Portfolio Standard Deviation')
plt.ylabel('Portfolio Mean Return')
plt.title('Efficient Frontier')
plt.show()
```

```python
# Maximum weight constraint (e.g., no more than 25% for each ETF) PART 4
max_weight = 0.25
bounds_with_max_weight = tuple((0, max_weight) for asset in range(n_assets))

# Re-run the optimization with the new constraints
optimal_variance_result_with_max_weight = minimize(portfolio_variance, initial_guess,
                                                   args=(covariance_matrix,), method='SLSQP',
                                                   bounds=bounds_with_max_weight, constraints=constraints)

# Function to calculate Sharpe Ratio
def sharpe_ratio(return_portfolio, std_portfolio, risk_free_rate=0.0173):
    return (return_portfolio - risk_free_rate) / std_portfolio

# Original Optimal Portfolio
original_weights = optimal_variance_result.x
original_portfolio_return = portfolio_mean(original_weights, annualized_mean)
original_portfolio_std = portfolio_std(original_weights, covariance_matrix)
original_sharpe_ratio = sharpe_ratio(original_portfolio_return, original_portfolio_std)
```

```python
# New Optimal Portfolio with Max Weight Constraint
new_weights = optimal_variance_result_with_max_weight.x
new_portfolio_return = portfolio_mean(new_weights, annualized_mean)
new_portfolio_std = portfolio_std(new_weights, covariance_matrix)
new_sharpe_ratio = sharpe_ratio(new_portfolio_return, new_portfolio_std)
```

```python
# Print the comparison PART 4
print("Comparison of Original and New Optimal Portfolios:")
print("--------------------------------------------------")
print(f"Original Portfolio Weights: {original_weights}")
print(f"New Portfolio Weights: {new_weights}")
print("--------------------------------------------------")
print(f"Original Portfolio Return: {original_portfolio_return}")
print(f"New Portfolio Return: {new_portfolio_return}")
print("--------------------------------------------------")
print(f"Original Portfolio Standard Deviation: {original_portfolio_std}")
print(f"New Portfolio Standard Deviation: {new_portfolio_std}")
print("--------------------------------------------------")
print(f"Original Portfolio Sharpe Ratio: {original_sharpe_ratio}")
print(f"New Portfolio Sharpe Ratio: {new_sharpe_ratio}")
```
