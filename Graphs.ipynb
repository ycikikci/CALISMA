```python
import plotly.graph_objs as go
import numpy as np

years = [2015, 2016, 2017, 2018, 2019, 2020]
class_population = [50, 55, 60, 65, 70, 75]

colors = ['rgb(0, 0, 255)', 'rgb(0, 128, 255)', 'rgb(0, 255, 255)', 'rgb(0, 255, 128)', 'rgb(0, 255, 0)', 'rgb(128, 255, 0)']

trace = go.Scatter(x=years, y=class_population, mode='lines+markers', marker=dict(color=colors, size=10), name='Class Population')

layout = go.Layout(title='Class Population Over Time', xaxis=dict(title='Year'), yaxis=dict(title='Population'), plot_bgcolor='black', paper_bgcolor='black', font=dict(color='white'))

fig = go.Figure(data=[trace], layout=layout)

fig.show()
```

```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020]
class_population = [50, 55, 60, 65, 70, 75]

colors = ['blue', 'cyan', 'lightblue', 'lightcyan', 'skyblue', 'deepskyblue']

plt.figure(figsize=(8, 8), facecolor='black')
plt.pie(class_population, labels=years, autopct='%1.1f%%', colors=colors)

plt.title('Class Population Distribution Over Time', fontsize=16, color='white')

plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020]
class_population = [50, 55, 60, 65, 70, 75]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(years, class_population, color='cyan', marker='o', linestyle='-', linewidth=2, label='Class Population')

ax.set_title('Class Population Over Time', fontsize=16, color='white')
ax.set_xlabel('Year', fontsize=14, color='white')
ax.set_ylabel('Population', fontsize=14, color='white')

fig.patch.set_facecolor('black')
ax.set_facecolor('black')

ax.grid(color='grey', linestyle='--', linewidth=0.5)

ax.legend(loc='lower right', fontsize=12, facecolor='black', edgecolor='white', labelcolor='white')

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020]
male_population = [25, 28, 30, 32, 35, 38]
female_population = [25, 27, 30, 33, 35, 37]

import pandas as pd
data = pd.DataFrame({'Year': years*2, 'Population': male_population + female_population, 'Sex': ['Male']*len(years) + ['Female']*len(years)})

sns.set_style("darkgrid")
sns.set_context("talk")

plt.figure(figsize=(10, 6), facecolor='none')
sns.barplot(data=data, x='Year', y='Population', hue='Sex', palette='muted')

plt.title('Class Population Distribution by Sex', fontsize=16, color='black')
plt.xlabel('Year', fontsize=14, color='black')
plt.ylabel('Population', fontsize=14, color='black')

plt.grid(color='purple', linestyle='--', linewidth=1)

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

turkey = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

turkey = turkey[turkey.name == 'Turkey']

city_names = ['Istanbul', 'Ankara', 'Izmir', 'Bursa', 'Antalya']
population_density = [2000, 1500, 1000, 800, 700]

cities = gpd.GeoDataFrame({'City': city_names, 'Population_Density': population_density},
                          geometry=gpd.points_from_xy([28.9784, 32.8597, 27.1428, 29.0765, 30.7133],
                                                    [41.0082, 39.9334, 38.4192, 40.1824, 36.8969]))

fig, ax = plt.subplots(figsize=(10, 8))
turkey.plot(ax=ax, color='lightgray', edgecolor='black')

cities.plot(ax=ax, column='Population_Density', cmap='viridis', legend=True, markersize=200,
            legend_kwds={'label': 'Population Density (persons/km²)', 'orientation': 'horizontal'})

for idx, row in cities.iterrows():
    ax.annotate(row['City'], xy=(row['geometry'].x, row['geometry'].y), xytext=(3, 3),
                textcoords="offset points", fontsize=8, color='black')

ax.set_title('Population Density in Cities of Turkey', fontsize=16)

#ax.axis('off')

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

thailand = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
thailand = thailand[thailand.name == 'Thailand']

city_names = ['Bangkok', 'Nonthaburi', 'Chiang Mai', 'Udon Thani', 'Pattaya']
population_density = [5200, 1500, 800, 600, 700]

cities = gpd.GeoDataFrame({'City': city_names, 'Population_Density': population_density},
                          geometry=gpd.points_from_xy([100.5167, 100.4733, 98.9817, 102.793, 100.88],  # Example longitudes
                                                    [13.7563, 13.8623, 18.7903, 17.4159, 12.9276]))  # Example latitudes

fig, ax = plt.subplots(figsize=(10, 8))
thailand.plot(ax=ax, color='lightgray', edgecolor='black')

cities.plot(ax=ax, column='Population_Density', cmap='viridis', legend=True, markersize=200,
            legend_kwds={'label': 'Population Density (persons/km²)', 'orientation': 'horizontal'})

for idx, row in cities.iterrows():
    ax.annotate(row['City'], xy=(row['geometry'].x, row['geometry'].y), xytext=(3, 3),
                textcoords="offset points", fontsize=8, color='black')

ax.set_title('Population Density in Cities of Thailand', fontsize=16)

ax.axis('off')

plt.show()
```
