```python
import pandas as pd
```

```python
internationals = pd.read_excel("International Investments on National Sectors.xlsx")
```

```python
credits = pd.read_excel("Credits_Sectors.xlsx")
```

```python
import matplotlib.pyplot as plt
credits.plot()
plt.xlabel("Sectors")
plt.ylabel("Thousands / TL")
plt.title("Credits of Sectors")
plt.show()
```

```python
internationals.plot()
plt.xlabel("Sectors")
plt.ylabel("Millions / USD")
plt.title("International Investments on National Sectors")
plt.show()
```

```python
internationals['2023 Ocak'] = internationals['2023 Ocak'].pct_change()
internationals['2023 Şubat'] = internationals['2023 Şubat'].pct_change()
internationals['2023 Mart'] = internationals['2023 Mart'].pct_change()
internationals['2023 Nisan'] = internationals['2023 Nisan'].pct_change()
internationals['2023 Mayıs'] = internationals['2023 Mayıs'].pct_change()
internationals['2023 Haziran'] = internationals['2023 Haziran'].pct_change()
internationals['2023 Temmuz'] = internationals['2023 Temmuz'].pct_change()
internationals['2023 Ağustos'] = internationals['2023 Ağustos'].pct_change()
internationals['2023 Eylül'] = internationals['2023 Eylül'].pct_change()
```

```python
internationals
```

```python
import numpy as np

internationals.replace([np.inf, -np.inf], np.nan, inplace=True)
internationals.dropna()
internationals
```

```python
merged_data = pd.merge(credits, internationals, how="inner", on=["Sektör Kodu"])
```

```python
merged_data
```

```python
avg_sectors_sept2023 = merged_data.groupby("Sektör Kodu")["2023 Eylül"].mean()
```

```python
avg_sectors_sept2023
```

```python
avg_credits_sept_2023 = merged_data.groupby("Sektör Kodu")["09_2023_Nakdi_Krediler"].mean()
```

```python
avg_credits_sept_2023
```

```python
merged_data2 = pd.merge(avg_sectors_sept2023, avg_credits_sept_2023, how="inner", on=["Sektör Kodu"])
```

```python
merged_data2
```

```python
#merged_data2 = merged_data2.pct_change().dropna()
#merged_data2
```

```python
merged_data2.columns=['investments','credits']
list(merged_data2.columns)
merged_data2
```

```python
import statsmodels.formula.api as smf

model = smf.ols("investments ~ credits", data=merged_data2)
results=model.fit()
print(results.summary())
```
