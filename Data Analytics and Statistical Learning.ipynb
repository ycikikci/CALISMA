a. Create a two random number generator rng1 and rng2 with the same seed value and

```python
import numpy as np

# Set the seed value
seed_value = 50

# Create two random number generators with the same seed
rng1 = np.random.default_rng(seed_value)
rng2 = np.random.default_rng(seed_value)

# Generate 10 standard normal random numbers from rng1
random_numbers_rng1 = rng1.standard_normal(size=10)

# Generate 10 standard normal random numbers from rng2
random_numbers_rng2 = rng2.standard_normal(size=10)

# Print the results
print("Random Numbers from rng1:", random_numbers_rng1)
print("Random Numbers from rng2:", random_numbers_rng2)
```

3. Using below code block (numpy and pandas packages) create a pair of x and monotonically 
increasing y values.

```python
import pandas as pd
import numpy as np

# Create x and y values
x = np.linspace(-10, 10, 101)
y = x ** 3 + 2 * x ** 2 + 5 * x + 1

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# a. Scatter plot
plt.scatter(df['x'], df['y'])
plt.title('Scatter Plot of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# b. Calculate Pearsonâ€™s R correlation coefficient between x and y
pearson_correlation = df.corr(method='pearson')['x']['y']
print(f"b. Pearson's R correlation coefficient between x and y: {pearson_correlation}")

# c. Convert x and y values to rank order values and calculate Pearsonâ€™s R correlation coefficient on these rank values
rank=df.rank(method='min')
print(rank)
rank_correlation=rank.corr(method='pearson')['x']['y']
print(f"c. Pearson's R correlation coefficient on rank values: {rank_correlation}")

# d. Now calculate Spearman rank correlation coefficient using corr method with method='spearman'
spearman_correlation = df[['x', 'y']].corr(method='spearman').iloc[0, 1]
print(f"d. Spearman rank correlation coefficient: {spearman_correlation}")
```

b) Pearsonâ€™s R correlation result is near to 1. There is a strong relation between x and y. 
c-d) Pearsonâ€™s R correlation coefficient on rank values and Spearman rank correlation coefficient are 1. That means they are perfectly correlated.

c. Using statsmodels package formula interface fit an OLS linear regression model. Predict 
values for the training and calculate errors. Answer d and e. 

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
results =smf.ols('y_with_error ~ x', data=df).fit()
results.summary()

df['predicted_y']=results.predict(df)
df['errors']=df['y_with_error']-df['predicted_y']
print(f"Predicted values for training:\n{df['predicted_y']}")
print(f"Calculated errors:\n{df['errors']}")
```
